#### while loop

```python
x = 1
while x < 4 :
    print(x)
    x = x + 1
```

```python
error = 50.0
while error > 1 :
    error = error / 4
    print(error)
```

```python
offset = 4

while offset != 0:
    print("correcting...")
    offset = offset - 1
    print(offset)
```

#### conditional while loop

```python
offset = -4

while offset != 0 :
    print("correcting...")
    if offset > 0:
        offset = offset - 1
    else:
        offset = offset + 1
    print(offset)
```

## _5. Case Study: Hacker Statistics_:
-  random number generator

#### generate random float

```python
np.random.seed(123)
print(np.random.rand())
```

#### generate random int dice roll

```python
print(np.random.randint(1,7))
print(np.random.randint(1,7))
```

#### construct random step

```python
np.random.seed(123)
step = 50
dice = np.random.randint(1,7)

if dice <= 2 :
    step = step - 1
elif dice <= 5:
    step = step + 1
else:
    step = step + np.random.randint(1,7)

print(dice)
print(step)
```

#### construct random walk

```python
np.random.seed(123)
random_walk = [0] # initialize random walk

for x in range(100) :
    step = random_walk[-1] # last element in random walk
    dice = np.random.randint(1,7)
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)

print(random_walk)
```

```python
np.random.seed(123)
random_walk = [0] # initialize random walk

for x in range(100) :
    step = random_walk[-1] # last element in random walk
    dice = np.random.randint(1,7)
    if dice <= 2:
        step = max(0,step - 1) # step can't go below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)

print(random_walk)
```

#### visualize walk

```python
plt.plot(random_walk)
plt.show()
```

#### simulate multiple walks

```python
np.random.seed(123)
all_walks = []

for i in range(10) : # simulates random walk 10 times
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    
    all_walks.append(random_walk)

print(all_walks)
```

#### visualize all walks

```python
np.random.seed(123)
all_walks = []

for i in range(10) : 
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    
    all_walks.append(random_walk)

np_aw = np.array(all_walks)

plt.plot(np_aw)
plt.show()
plt.clf()

np_aw_t = np.transpose(np_aw)

plt.plot(np_aw_t)
plt.show()
```

#### implement possible reset

```python
np.random.seed(123)
all_walks = []

for i in range(250) :
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # if value is less than or equal to 0.001 reset step to 0
        if np.random.rand() <= 0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
np.random.seed(123)
all_walks = []

for i in range(500) : # simulates random walk 500 times
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
ends = np_aw_t[-1]

plt.hist(ends)
plt.show()
```
